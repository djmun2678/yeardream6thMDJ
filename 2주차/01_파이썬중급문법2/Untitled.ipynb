```python
class TeamMember:
    """
    IT 회사 팀원을 표현하는 클래스
    
    속성:
        name (str): 팀원 이름
        role (str): 직책 (예: 시니어 개발자, 주니어 기획자 등)
        salary (int): 연봉 (만원 단위)
    """
    
    pass # 코드 

    def __init__(self, name, role, salary):
        
        #self.속성명 = 값 => 인스턴스의 속성을 설정
        self.name = name
        self.role = role
        self.salary = salary
        self.projects = []  # 현재 참여 중인 프로젝트 목록

```

```python
alice = TeamMember(name = "Alice", role = "시니어 개발자", salary=6000)
bob = TeamMember(name = "Bob", role = "주니어 기획자", salary=3800)

# 속성에 접근 : 인스턴스. 속성명
print(f"{alice.name}의 직책: {alice.role}, 연봉: {alice.salary}만원")
print(f"{bob.name}의 직책: {bob.role}, 연봉: {bob.salary}만원")

```
