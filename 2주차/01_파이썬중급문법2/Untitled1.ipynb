```python
class TeamMember:

    def __init__(self, name, role):
        self.name = name
        self.role = role
```

```python
member1 = TeamMember("동진", "개발자")

print(member1.name)
print(member1.role)
```

```python
class TeamMember:

    def __init__(self, name, role, salary):
        self.name = name
        self.role = role
        self.salary = salary
    
```

```python
member1 = TeamMember("동진", "개발자", 5000)

print(member1.name)
print(member1.role)
print(member1.salary)
```
